There is no relation between those fields that we can use a field to determine the null values

50% of values of every column is near mean value. Thats why we can use mean value to fill those null value

```python
data.isnull().sum()
```

```python
#Filling the null values
data['ph']=data['ph'].fillna(data['ph'].mean())
data['Sulfate']=data['Sulfate'].fillna(data['Sulfate'].mean())
data['Trihalomethanes']=data['Trihalomethanes'].fillna(data['Trihalomethanes'].mean())
```

```python
#Again checking the null values
data.isnull().sum()
```

```python
#checking the ratio of potable and non-potable
data['Potability'].value_counts()
```

Data processing

```python
#Declaring the dependent and independent variable
x=data.drop('Potability',axis=1)
y=data['Potability']
```

```python
#importing libraries again for model selection
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifer
```

```python
#Dividing training and testing data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
#declaring the linear model
rfc=RandomForestClassifer ()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
ypred=rfc.predict(xtest)
```

```python
#accuracy of the model
accuracy=rfc.score(xtest,ytest)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,ypred)
score
```
